```python
# Store each projects ROI (Return of Investment) ATH
roi = []

for r in soup.find_all('div', attrs ={'class': 'data ng-star-inserted'}):
    if len(r.text) < 7:
        roi.append(r.text.strip())
        
# Turns out 5 projects have no ROI listed - generally because the project has not been listed to the public at the moment
# of gathering data. That is why I am not going to take these projects under consideration
roi_not_listed = ['chainport', 'forward', 'street-runner', 'launchx', 'openpredict']


# Needed by BeatifulSoup
# These are the names that need to be changed, so the URL on the daomaker.com website fits the names
missing = {'dragonsb': 'dragon-sb', 'meta-gaming-guild': 'mgg', 'three-kingdoms': '3kingdoms',
          'heroes-&-empires': 'heroesempires', 'fear': 'fear-nfts', 'epik-prime': 'epik',
          'makiswap': 'maki-swap', 'orao': 'orao-network', 'anrkey-x™': 'anrkey-x',
          'forward-protocol': 'forward'}
        
```

```python
# Change whitespaces to dashes/dot to fit the URLs
# Lowercase the names
projects = list(map(lambda x: x.replace(' ', '-'), projects))
projects = list(map(lambda x: x.replace('.', '-'), projects))
projects = list(map(lambda x: x.lower(), projects))

projects = [x if x not in missing else missing[x] for x in projects]


df = pd.DataFrame(data = raised, index = projects, columns=['raised'])
# Dropping projects without ROI
df = df.drop(roi_not_listed)

df['ROI_ATH'] = roi
```

```python
# Loop through each project's summary and get data
for name in df.index:
    fp = requests.get('https://daomaker.com/company/' + name)
    s = BeautifulSoup(fp.text)
    
    # CATEGORIES FILLING
    cat = s.find('span', attrs={'class': 'medium_tag grey_bg'})
    
    df.loc[name, 'catg'] = cat.text
        
        
    # OTHER COMPANIES BEHIND THIS PROJECT
        
    # Is there Venture Capital and Market Maker behind the project?
    if len(s.find_all('h4', attrs={'class': 'comp_name'})[0].text) == 0:
        df.loc[name, 'vc'] = np.nan
    else:
        df.loc[name, 'vc'] = s.find_all('h4', attrs={'class': 'comp_name'})[0].text
    
    # Is there Market Maker?
    if len(s.find_all('h4', attrs={'class': 'comp_name'})[1].text) == 0:
        df.loc[name, 'mm'] = np.nan
    else:
        df.loc[name, 'mm'] = s.find_all('h4', attrs={'class': 'comp_name'})[1].text
    
    # Is cap controlled?
    if 'grey' in str(s.find_all('div', attrs={'class': 'image_wrap'})[-4]):
        df.loc[name, 'cap_controlled'] = 0
    else:
        df.loc[name, 'cap_controlled'] = 1
    
    # Is approved by DAO?
    if 'grey' in str(s.find_all('div', attrs={'class': 'image_wrap'})[-3]):
        df.loc[name, 'dao_approve'] = 0
    else:
        df.loc[name, 'dao_approve'] = 1     

    l = []
    for t in s.find('div', attrs={'id': 'company-content'}):
        l.append(t.text)
    df.loc[name, 'desc'] = ' '.join(l)     
    
```

```python
# Get Coingecko ids for every project so I can acces Coingecko API easily
for name in df.index:
    
    fp = requests.get('https://coingecko.com/en/coins/' + name)
    s = BeautifulSoup(fp.text)
    
    id_ = s.find('div', attrs = {'class': "tw-px-2.5 tw-py-1.5 tw-my-0.5 tw-mr-1 tw-rounded-md tw-text-sm tw-font-medium tw-bg-gray-100 tw-text-gray-800 hover:tw-bg-gray-200 dark:tw-text-white dark:tw-bg-white dark:tw-bg-opacity-10 dark:hover:tw-bg-opacity-20 dark:focus:tw-bg-opacity-20"})
    if id_ != None:
        df.loc[name, 'id'] = id_.text.strip()
```

```python
# Need to change some ids manually and update the dataframe
different_id = {'dragon-sb': 'dragonsb', 'mgg': 'metagaming-guild', 'meta-soccer': 'metasoccer',
               'kaka-nft': 'kaka-nft-world', 'polygonum-online': 'pog-coin', 'resource-finance': 'resource-protocol',
               'envoy': 'envoy-network', '3kingdoms': 'the-three-kingdoms',
               'heroesempires': 'heroes-empires', 'lossless-protocol': 'lossless', 'fear-nfts': 'fear',
               'epik': 'epik-prime', 'maki-swap': 'makiswap', 'legends-of-crypto': 'locgame',
               'crypto-prophecies': 'the-crypto-prophecies', 'smoothy-finance': 'smoothy',
               'mochi': 'mochi-market', 'showcase': 'showcase-token', 'crowny': 'crowny-token', 'aluna-social': 'aluna',
               'seascape': 'crowns', 'prometheus': 'metagame-arena'}

for name in df.index:
    if name in different_id.keys():
        df.loc[name, 'id'] = different_id[name]
```

```python
# Acess Coingecko API
cg = CoinGeckoAPI()

for index, row in df.iterrows():
    # Sleep so we avoid Coingecko's limitation on 50accesses per minute limiation
    time.sleep(2.0)
    
    # Get timestamp when the project was publicly listed
    milis = cg.get_coin_market_chart_by_id(row.id, vs_currency='usd', days='max')['prices'][0][0]
    # Transform timestamp to readable date
    df.loc[index, 'launch_date'] = datetime.fromtimestamp(milis/1000.0).strftime('%d/%m/%y')
    
    # Get the general info about the coin
    data = cg.get_coin_by_id(row.id)    
    
    df.loc[index,'platform'] = data['asset_platform_id']
    df.loc[index, 'total_supply'] = data['market_data']['total_supply']

        
```

```python
url = requests.get('https://api.alternative.me/fng/?limit=0&format=json')
data = json.loads(url.text)

# Get the fear and greed index of Bitcoin on the day that project was publicly listed
for d in data['data']:
    t = d['timestamp']
    date = datetime.fromtimestamp(int(t)).strftime('%d/%m/%y')
    project = df[df['launch_date'] == date].index
    df.loc[project, 'fear_greed_val'] = d['value']
```
